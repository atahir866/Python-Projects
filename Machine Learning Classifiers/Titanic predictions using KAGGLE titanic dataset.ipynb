```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn import svm
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.preprocessing import MinMaxScaler
```

```python
train=pd.read_csv('training data.csv')
train.head()
```

```python
# Making additional features

train['sibsp_parch']=train['SibSp']+train['Parch']
train['Age*Class']=train['Age']*train['Pclass']
a=[]
n=[]
for x in train['Sex']:
    if x == 'male':
        a.append(1)
    elif x == 'female':
        a.append(0)
train['gender']=a

for m in train['Embarked']:
    if m=='S':
        n.append(0)
    elif m=='C':
        n.append(1)
    elif m=='Q':
        n.append(2)
    else:
        n.append(3)
train['begin_port']=n
train['Fare_Per_Person']=train['Fare']/(train['sibsp_parch']+1)
```

```python
# handling missing values

train = train.fillna(method='ffill')
train.head()
```

```python
# splitting the training data into test and train data to check accuracy because we dont have test data for the dataset

X=train[['Pclass','gender','sibsp_parch','Age*Class','begin_port','Fare_Per_Person']]
y=train['Survived']
```

```python
scaler=MinMaxScaler()
xtrain_scaled=scaler.fit_transform(X)
```

```python
# creating objects of different classifiers

alt=svm.SVC(gamma=2, C=1)
clf=DecisionTreeClassifier()
knn = KNeighborsClassifier(n_neighbors = 4,weights='uniform')
rnd=RandomForestClassifier()
gas=GaussianNB()
ada=AdaBoostClassifier(n_estimators=1000,learning_rate=0.1)
mlp=MLPClassifier(activation='relu',learning_rate='invscaling')

```

```python
# choosing the classifier with the best accuracy after cross validation

scores = cross_val_score(ada, xtrain_scaled, y, cv=50,scoring='f1_macro')
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
test=pd.read_csv('testing data.csv')
test = test.fillna(method='ffill')
test.head()
```

```python
# Making additional features in testing data and handling missing values

test['sibsp_parch']=test['SibSp']+test['Parch']
test['Age*Class']=test['Age']*test['Pclass']
a=[]
for x in test['Sex']:
    if x == 'male':
        a.append(1)
    elif x == 'female':
        a.append(0)
test['gender']=a
n=[]
for m in test['Embarked']:
    if m=='S':
        n.append(0)
    elif m=='C':
        n.append(1)
    elif m=='Q':
        n.append(2)
    else:
        n.append(3)
len(n)
test['begin_port']=n
test['Fare_Per_Person']=test['Fare']/(test['sibsp_parch']+1)

test = test.fillna(method='ffill')
test.head()
```

```python
X_test=test[['Pclass','gender','sibsp_parch','Age*Class','begin_port','Fare_Per_Person']]
xtest_scaled=scaler.transform(X_test)
```

```python
# training Ada boost classifier

ada.fit(xtrain_scaled,y)
```

```python
# Using ada boost classifier to predict on testing data
a=ada.predict(xtest_scaled)
```

```python
# putting passenger ID and predictions into a Pandas Dataframe
p=test['PassengerId']
result=pd.DataFrame(p)
result['Survived']=a
```

```python
# exporting dataframe to CSV 
result.to_csv('results.csv', sep=',',encoding='utf-8')
```
